```python
# All projects names that were missing have been filled using project codes

import numpy as np

data= json.load((open('C:/Users/Owner/Documents/data_science/1574117884_data_wrangling_json_5_/data_wrangling_json/data/World_Bank_projects.json')))
normalized_data = json_normalize(data, record_path='mjtheme_namecode')
df = pd.DataFrame(normalized_data)
new_df = (df.replace({'name':{'':np.nan}})
          .sort_values(['code', 'name'])
          .fillna(method='ffill')
          .sort_index())
print(new_df)
```
